```python
df=df.drop(["Unnamed: 0", "Enroll"], axis=1)
```

```python
df.Private = df.Private.replace("Yes",1)
df.Private = df.Private.replace("No",0)
```

```python
df_norm = (df - df.min(axis=0)) / (df.max(axis=0) - df.min(axis=0))
```

```python
df_norm
```

```python
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters = 3, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(df_norm)
```

```python
kmeans.cluster_centers_
```

```python
df["Clusters"]=kmeansmodel.labels_
df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_df = pca.fit_transform(df_norm)
pca_df_df = pd.DataFrame(data=pca_df,columns =["Component_1","Component_2"])
pca_new_df = pd.concat([pca_df_df,df[["Clusters"]]],axis=1)

pca_new_df
```

```python
fig = plt.figure(figsize=(6,6))

ax = fig.add_subplot(1,1,1)

ax.set_xlabel("Component_1",fontsize=15)
ax.set_ylabel("Component_2",fontsize=15)
ax.set_title("Clusters of Student")

color_theme = np.array(["blue","green","orange"])

ax.scatter(x=pca_new_df.Component_1, y=pca_new_df.Component_2, c=color_theme[pca_new_df.Clusters], s = 50)

plt.show()
```
